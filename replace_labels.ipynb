```python
import pandas as pd
import os

# Define the file path
# Assuming the notebook is in the root of the repository
file_path = r"(A) Data/Need To Vote/News Title_2000 Data.csv"

# If running locally and the path isn't found, try the absolute path (commented out by default or handled via check)
absolute_path = r"c:\Users\Asus\Documents\GitHub\News-Scrapping\(A) Data\Need To Vote\News Title_2000 Data.csv"

if os.path.exists(file_path):
    load_path = file_path
else:
    load_path = absolute_path
    
print(f"Loading data from: {load_path}")

df = pd.read_csv(load_path)

# Display first 5 rows
display(df.head())
```

```python
# Check for null values before dropping
print("Shape before dropping NaNs:", df.shape)
print("Null values counts:\n", df.isnull().sum())

# Drop rows with NaN values
df = df.dropna()

# Verify after dropping
print("\nShape after dropping NaNs:", df.shape)
print("Null values counts:\n", df.isnull().sum())
```

```python
# Check unique values in 'Labelling' column before replacement
print("Unique values before replacement:", df['Labelling'].unique())
```

```python
# Define the mapping dictionary
label_mapping = {
    'Negative': -1,
    'Neutral': 0,
    'Positive': 1
}

# Replace the values
df['Labelling'] = df['Labelling'].replace(label_mapping)

# Verify the replacement
print("Unique values after replacement:", df['Labelling'].unique())
display(df.head())
```

```python
# Save the processed data to a new CSV file to preserve the original
output_path = load_path.replace(".csv", "_Labelled.csv")
df.to_csv(output_path, index=False)

print(f"File saved to: {output_path}")
```
